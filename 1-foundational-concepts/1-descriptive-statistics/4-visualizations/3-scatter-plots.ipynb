## Scatter Plots
**Purpose**: Show relationships between two numerical variables

**Statistical Use:**
- Identify correlations (positive, negative, none) and patterns
- Detect outliers
- Explore potential causal relationships

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data: Study hours vs Exam scores
study_hours = np.random.normal(15, 5, 50)
exam_scores = 50 + 2.5 * study_hours + np.random.normal(0, 8, 50)

plt.figure(figsize=(10, 6))
plt.scatter(study_hours, exam_scores, alpha=0.6, color='green')
plt.xlabel('Study Hours per Week')
plt.ylabel('Exam Score')
plt.title('Study Hours vs Exam Scores')
plt.grid(True, alpha=0.3)

# Add correlation line
z = np.polyfit(study_hours, exam_scores, 1)
p = np.poly1d(z)
plt.plot(study_hours, p(study_hours), "r--", alpha=0.8, label='Trend line')

plt.legend()
plt.show()
```
