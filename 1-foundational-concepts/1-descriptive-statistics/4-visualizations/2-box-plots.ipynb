## Box Plots
**Purpose**: Visualize data distribution through quartiles and identify outliers

**Statistical Use:**
- Show median, quartiles, and outliers
- Compare distributions across groups
- Identify skewness and spread

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Sample data for different groups
data = {
    'Group A': np.random.normal(100, 15, 100),
    'Group B': np.random.normal(110, 12, 100),
    'Group C': np.random.normal(95, 20, 100)
}
df = pd.DataFrame(data)

plt.figure(figsize=(8, 6))
sns.boxplot(data=df)
plt.ylabel('Scores')
plt.title('Comparison of Scores Across Groups')
plt.show()

# Key elements shown:
# - Median (middle line)
# - IQR (box)
# - Outliers (dots beyond whiskers)
```
