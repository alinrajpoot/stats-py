# Histograms
**Purpose**: Show frequency distribution of numerical data

**Statistical Use:**
- Understand data distribution shape
- Identify central tendency, spread, skewness
- Spot outliers and gaps



```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
heights = np.random.normal(170, 10, 1000)  # Mean=170cm, SD=10

plt.figure(figsize=(10, 4))
plt.hist(heights, bins=30, alpha=0.7, color='skyblue', edgecolor='black')
plt.axvline(heights.mean(), color='red', linestyle='--', label=f'Mean: {heights.mean():.1f}cm')
plt.xlabel('Height (cm)')
plt.ylabel('Frequency')
plt.title('Distribution of Heights')
plt.legend()
plt.show()
```
