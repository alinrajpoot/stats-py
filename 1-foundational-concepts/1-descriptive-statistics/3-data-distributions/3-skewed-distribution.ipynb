## Skewed Distribution
A skewed distribution is a probability distribution that is not symmetrical around its mean. In a skewed distribution, the tail on one side of the distribution is longer or fatter than the other side. Skewness can be positive (right-skewed) or negative (left-skewed).

- Right Skew (Positive Skew):
  - Tail on the right side is longer
  - Mean > Median > Mode
  - Most data clustered on the left
  - Example: Income distribution

- Left Skew (Negative Skew):
  - Tail on the left side is longer
  - Mean < Median < Mode
  - Most data clustered on the right
  - Example: Age at retirement

Real-world examples:

- Income distribution
- Real estate prices
- Customer purchase behavior

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import skewnorm

# Create skewed distributions
np.random.seed(42)

# Right-skewed (positive skew)
right_skewed = skewnorm.rvs(5, loc=50, scale=20, size=1000)  # Positive skew parameter

# Left-skewed (negative skew)  
left_skewed = skewnorm.rvs(-5, loc=50, scale=20, size=1000)  # Negative skew parameter

# Plot comparison
fig, axes = plt.subplots(2, 2, figsize=(12, 8))

# Right-skewed plots
axes[0, 0].hist(right_skewed, bins=30, alpha=0.7, color='lightcoral', edgecolor='black')
axes[0, 0].set_title('Right-Skewed Distribution')
axes[0, 0].set_xlabel('Values')
axes[0, 0].set_ylabel('Frequency')

axes[0, 1].boxplot(right_skewed, vert=False)
axes[0, 1].set_title('Right-Skewed - Box Plot')

# Left-skewed plots
axes[1, 0].hist(left_skewed, bins=30, alpha=0.7, color='lightgreen', edgecolor='black')
axes[1, 0].set_title('Left-Skewed Distribution')
axes[1, 0].set_xlabel('Values')
axes[1, 0].set_ylabel('Frequency')

axes[1, 1].boxplot(left_skewed, vert=False)
axes[1, 1].set_title('Left-Skewed - Box Plot')

plt.tight_layout()
plt.show()

# Statistical analysis
def analyze_skewness(data, name):
    from scipy.stats import skew as statskew
    mean = np.mean(data)
    median = np.median(data)
    skew = statskew(data)
    
    print(f"\n{name} Analysis:")
    print(f"Mean: {mean:.2f}")
    print(f"Median: {median:.2f}")
    print(f"Skewness: {skew:.4f}")
    print(f"Mean > Median: {mean > median}")
    
    if skew > 0:
        print("Distribution is RIGHT-SKEWED")
    elif skew < 0:
        print("Distribution is LEFT-SKEWED")
    else:
        print("Distribution is SYMMETRICAL")

analyze_skewness(right_skewed, "Right-Skewed Data")
analyze_skewness(left_skewed, "Left-Skewed Data")
```
