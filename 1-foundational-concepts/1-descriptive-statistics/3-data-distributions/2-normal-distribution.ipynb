## Normal Distribution (Gaussian Distribution)
A normal distribution is a continuous probability distribution characterized by its bell-shaped curve, which is symmetric around the mean. It is defined by two parameters: the mean (μ) and the standard deviation (σ). The mean determines the center of the distribution, while the standard deviation measures the spread or dispersion of the data points around the mean.

- Bell-shaped curve
- Symmetrical around the mean
- Mean = Median = Mode
- Follows the 68-95-99.7 rule:
  - 68% of data within 1σ of mean
  - 95% of data within 2σ of mean
  - 99.7% of data within 3σ of mean

Real-world examples:

- Heights of people
- Test scores
- Measurement errors
- Blood pressure readings

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
from scipy.stats import norm, skewnorm

# Generate normal data
np.random.seed(42)
normal_data = np.random.normal(loc=50, scale=15, size=1000)  # mean=50, std=15

# Plot normal distribution
plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
plt.hist(normal_data, bins=30, density=True, alpha=0.7, color='skyblue', edgecolor='black')
plt.title('Normal Distribution Histogram')
plt.xlabel('Values')
plt.ylabel('Density')

plt.subplot(1, 3, 2)
# Probability Density Function (PDF)
x = np.linspace(normal_data.min(), normal_data.max(), 100)
pdf = norm.pdf(x, normal_data.mean(), normal_data.std())
plt.plot(x, pdf, 'r-', linewidth=2, label='PDF')
plt.hist(normal_data, bins=30, density=True, alpha=0.5, color='lightblue')
plt.title('Normal Distribution PDF')
plt.xlabel('Values')
plt.ylabel('Density')
plt.legend()

plt.subplot(1, 3, 3)
# Q-Q plot for normality check
stats.probplot(normal_data, dist="norm", plot=plt)
plt.title('Q-Q Plot - Normality Check')

plt.tight_layout()
plt.show()

# Statistical tests for normality
from scipy.stats import shapiro, normaltest

print("Normality Tests:")
print(f"Shapiro-Wilk p-value: {shapiro(normal_data)[1]:.4f}")
print(f"Normaltest p-value: {normaltest(normal_data)[1]:.4f}")
print(f"Mean: {normal_data.mean():.2f}, Median: {np.median(normal_data):.2f}")
print(f"Skewness: {stats.skew(normal_data):.4f}")
```
